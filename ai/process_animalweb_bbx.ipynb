```python
# ==========================================================
# ライブラリのインポート
# ==========================================================
import os
import glob
import cv2
import numpy as np
from tqdm import tqdm
import shutil
from google.colab import files  # ローカル保存用

# ==========================================================
# 設定
# ==========================================================
BASE_DIR = "/content/drive/MyDrive/animalweb_project/raw/animal_dataset_v1_clean_check"
WORK_DIR = "/content/animalweb_work"  # 一時作業ディレクトリ
OUTPUT_DIR = os.path.join(WORK_DIR, "jpg_pts_dataset")
os.makedirs(OUTPUT_DIR, exist_ok=True)

IMG_SIZE = 224
padding_ratio = 0.3  # 余白の割合

# ==========================================================
# .pts読み込み関数
# ==========================================================
def read_pts(path):
    with open(path, "r") as f:
        lines = f.readlines()
    pts = []
    read = False
    for line in lines:
        line = line.strip()
        if line == "{":
            read = True
            continue
        if line == "}":
            break
        if read:
            try:
                x, y = map(float, line.split())
                pts.append([x, y])
            except:
                continue
    return np.array(pts)

# ==========================================================
# 画像リサイズ関数
# ==========================================================
def resize_img(im):
    old_size = im.shape[:2]  # (height, width)
    ratio = float(IMG_SIZE) / max(old_size)
    new_size = tuple([int(x * ratio) for x in old_size])
    im = cv2.resize(im, (new_size[1], new_size[0]))
    delta_w = IMG_SIZE - new_size[1]
    delta_h = IMG_SIZE - new_size[0]
    top, bottom = delta_h // 2, delta_h - (delta_h // 2)
    left, right = delta_w // 2, delta_w - (delta_w // 2)
    new_im = cv2.copyMakeBorder(im, top, bottom, left, right,
                                cv2.BORDER_CONSTANT, value=[0, 0, 0])
    return new_im, ratio, top, left

# ==========================================================
# .ptsファイルを探索
# ==========================================================
pts_files = sorted(glob.glob(os.path.join(BASE_DIR, "*.pts")))
print(f"発見したサンプル数: {len(pts_files)}")

# ==========================================================
# 各画像を保存（jpg + bb.pts）
# ==========================================================
for pts_path in tqdm(pts_files, desc="Processing"):
    name = os.path.splitext(os.path.basename(pts_path))[0]
    img_path = os.path.join(BASE_DIR, name + ".jpg")
    if not os.path.exists(img_path):
        continue

    # ランドマーク読み込み
    landmarks = read_pts(pts_path)
    if landmarks.size == 0:
        continue

    # 画像読み込み
    img = cv2.imread(img_path)
    if img is None:
        continue

    # リサイズ＆座標変換
    img, ratio, top, left = resize_img(img)
    landmarks = ((landmarks * ratio) + np.array([left, top])).astype(np.int32)

    # バウンディングボックス計算（余白30%固定）
    x_min, y_min = landmarks[:, 0].min(), landmarks[:, 1].min()
    x_max, y_max = landmarks[:, 0].max(), landmarks[:, 1].max()
    w, h = x_max - x_min, y_max - y_min
    pad_w, pad_h = int(w * padding_ratio), int(h * padding_ratio)

    x_min = max(0, x_min - pad_w)
    y_min = max(0, y_min - pad_h)
    x_max = min(img.shape[1]-1, x_max + pad_w)
    y_max = min(img.shape[0]-1, y_max + pad_h)

    # --- 画像保存 ---
    cv2.imwrite(os.path.join(OUTPUT_DIR, f"{name}.jpg"), img)

    # --- バウンディングボックス保存 (.pts形式) ---
    bb_pts_path = os.path.join(OUTPUT_DIR, f"{name}.pts")
    with open(bb_pts_path, "w") as f:
        f.write("version: 1\n")
        f.write("{\n")
        f.write(f"{x_min} {y_min}\n")
        f.write(f"{x_max} {y_max}\n")
        f.write("}\n")

print(f"✅ 保存完了: {len(os.listdir(OUTPUT_DIR))} files")

# --- ZIP圧縮 ---
zip_path = os.path.join(WORK_DIR, "animalweb_jpg_pts_dataset.zip")
shutil.make_archive(zip_path.replace(".zip", ""), "zip", OUTPUT_DIR)
print(f"✅ ZIP圧縮完了: {zip_path}")

# --- ローカルにダウンロード ---
files.download(zip_path)
```

```python
## 📦 データセットの中身確認コード

import os
import glob
import cv2
import numpy as np
import random
import matplotlib.pyplot as plt

# ==========================================================
# 設定
# (元のスクリプトの設定を引き継ぎます)
# ==========================================================
WORK_DIR = "/content/animalweb_work"
OUTPUT_DIR = os.path.join(WORK_DIR, "jpg_pts_dataset")

# ==========================================================
# .pts読み込み関数（元のスクリプトから再利用）
# ==========================================================
def read_pts(path):
    with open(path, "r") as f:
        lines = f.readlines()
    pts = []
    read = False
    for line in lines:
        line = line.strip()
        if line == "{":
            read = True
            continue
        if line == "}":
            break
        if read:
            try:
                # バウンディングボックスのptsファイルは(x_min, y_min)と(x_max, y_max)の2点です
                x, y = map(float, line.split())
                pts.append([x, y])
            except:
                continue
    return np.array(pts)

# ==========================================================
# データセットの中身を確認
# ==========================================================

# 出力ディレクトリ内の.jpgファイルを全て取得
jpg_files = glob.glob(os.path.join(OUTPUT_DIR, "*.jpg"))
print(f"確認対象の画像ファイル数: {len(jpg_files)}")

if not jpg_files:
    print("⚠️ データセットディレクトリに画像ファイルが見つかりませんでした。処理が正しく行われたか確認してください。")
else:
    # ランダムに表示するサンプル数
    num_samples_to_check = min(5, len(jpg_files))

    # サンプルをランダムに選択
    selected_files = random.sample(jpg_files, num_samples_to_check)

    plt.figure(figsize=(15, 5 * num_samples_to_check))

    for i, img_path in enumerate(selected_files):
        name = os.path.splitext(os.path.basename(img_path))[0]
        pts_path = os.path.join(OUTPUT_DIR, f"{name}.pts")

        # 1. 画像を読み込み
        img = cv2.imread(img_path)
        # BGRをRGBに変換 (matplotlibで表示するため)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # 2. バウンディングボックス座標を読み込み
        bb_pts = read_pts(pts_path)

        # 3. 画像にバウンディングボックスを描画
        if bb_pts.shape == (2, 2):
            # 座標は [x_min, y_min] と [x_max, y_max] の2点
            x_min, y_min = map(int, bb_pts[0])
            x_max, y_max = map(int, bb_pts[1])

            # バウンディングボックスの描画 (色は緑、太さ2)
            cv2.rectangle(img_rgb, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
            title = f"{name}.jpg\nBB: ({x_min},{y_min}) to ({x_max},{y_max})"
        else:
            title = f"{name}.jpg\n⚠️ .ptsファイル形式エラーまたはデータなし"

        # 4. 表示
        plt.subplot(num_samples_to_check, 1, i + 1)
        plt.imshow(img_rgb)
        plt.title(title)
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```
